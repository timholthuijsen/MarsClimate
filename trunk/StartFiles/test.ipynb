```python
import netCDF4 as nc
```

```python
ModelData = 'start_archive_64x48x49_MY24.nc'
reforecast = nc.Dataset(ModelData)
```

```python
print(reforecast)
```

```python
lons = reforecast.variables['longitude'][:]
lats = reforecast.variables['latitude'][:]
```

```python
temp = reforecast.variables['temp'][0,0,:,:]


temp_units = reforecast.variables['temp'].units
```

```python
temp
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

plt.figure(figsize=(11,8.8))
ax = plt.axes(projection=ccrs.PlateCarree())

plt.contourf(lons, lats, temp, 60, cmap = 'coolwarm',
             transform=ccrs.PlateCarree())

ax.coastlines()

plt.show()
```
